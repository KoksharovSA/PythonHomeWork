```python
import pandas as pd
from sklearn import model_selection, ensemble, linear_model, metrics 
```

```python
df = pd.read_csv('diamonds.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
dict_cut = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
dict_color = {'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7}
dict_clarity = {'I1': 1, 'SI2': 2, 'SI1': 3, 'VS2': 4, 'VS1': 5, 'VVS2': 6, 'VVS1': 7, 'IF': 8}
```

```python
df['cut'] = df['cut'].map(dict_cut)
df['color'] = df['color'].map(dict_color)
df['clarity'] = df['clarity'].map(dict_clarity)
```

```python
df.head(10)
```

```python
split = int(df.shape[0]*0.8)
df_train = df.iloc[:split]
df_test = df.iloc[split:]
print(df.shape[0])
print(df_train.shape)
print(df_test.shape)
print(df_train.shape[0]+df_test.shape[0])
```

```python
x_df_train = df_train.drop(['price'], axis=1)
y_df_train = df_train['price']
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_df_train, y_df_train, test_size=0.2, random_state=42)
```

```python
# reg = ensemble.RandomForestRegressor(max_depth=2, random_state=0)
# reg = linear_model.BayesianRidge()
# reg = linear_model.Lasso(alpha=0.1)
reg = linear_model.Ridge(alpha=.5)
reg.fit(x_train, y_train)
test_y_predict_reg = reg.predict(x_test)
ra = round(metrics.explained_variance_score(y_test, test_y_predict_reg, multioutput='uniform_average'),6)
print(ra)
```
